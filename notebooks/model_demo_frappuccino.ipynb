# Starbucks Frappuccino Optimizer — Model Demo

This notebook demonstrates how to load and use the published Kaggle Model.

```python
import pickle
from pathlib import Path
import os

# ── Load model from Kaggle Model ─────────────────────────────────────
candidates = [
    Path('/kaggle/input/models/shiratoriseto/starbucks-frappuccino-optimizer/other/default/1/frappuccino_optimizer.pkl'),
    Path('/kaggle/input/starbucks-frappuccino-optimizer/other/default/1/frappuccino_optimizer.pkl'),
    Path('../models/frappuccino_optimizer.pkl'),
]

model_path = None
for p in candidates:
    if p.exists():
        model_path = p
        break

if model_path is None:
    for root in ['/kaggle/input/models', '/kaggle/input']:
        if Path(root).exists():
            for dirpath, dirnames, filenames in os.walk(root):
                for f in filenames:
                    print(os.path.join(dirpath, f))
    raise FileNotFoundError('Model not found')

with open(model_path, 'rb') as f:
    model = pickle.load(f)

print(f'Model: {model["name"]} v{model["version"]}')
print(f'Keys: {list(model.keys())}')
```

```python
# ── Scoring function details ─────────────────────────────────────────
print('=== Scoring Function ===')
print(f'Formula: {model["scoring_function"]["formula"]}')
print(f'\nDesirability components:')
for k, v in model['scoring_function']['desirability'].items():
    print(f'  {k}: {v}')
print(f'\nAffordability:')
for k, v in model['scoring_function']['affordability'].items():
    print(f'  {k}: {v}')
```

```python
# ── Optimization bounds ──────────────────────────────────────────────
import pandas as pd

bounds = model['optimization']['bounds']
bounds_df = pd.DataFrame([
    {'parameter': k, 'min': v[0], 'max': v[1]} for k, v in bounds.items()
])
print(f'Method: {model["optimization"]["method"]}')
print(f'\nParameter Bounds:')
bounds_df
```

```python
# ── April 2026 optimal product ───────────────────────────────────────
optimal = model['april_2026_optimal']
print('=== April 2026 Optimal New Frappuccino ===')
for k, v in optimal.items():
    print(f'  {k}: {v}')
```
