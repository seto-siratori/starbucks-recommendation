# Starbucks Recommendation Engine — Model Demo

This notebook demonstrates how to load and use the published Kaggle Model.

```python
import pickle
from pathlib import Path
import os

# ── Load model from Kaggle Model ─────────────────────────────────────
candidates = [
    Path('/kaggle/input/models/shiratoriseto/starbucks-recommendation-engine-model/other/default/1/recommendation_engine.pkl'),
    Path('/kaggle/input/starbucks-recommendation-engine-model/other/default/1/recommendation_engine.pkl'),
    Path('../models/recommendation_engine.pkl'),
]

model_path = None
for p in candidates:
    if p.exists():
        model_path = p
        break

if model_path is None:
    # Debug: list available paths
    for root in ['/kaggle/input/models', '/kaggle/input']:
        if Path(root).exists():
            for dirpath, dirnames, filenames in os.walk(root):
                for f in filenames:
                    print(os.path.join(dirpath, f))
    raise FileNotFoundError('Model not found')

with open(model_path, 'rb') as f:
    model = pickle.load(f)

print(f'Model: {model["name"]} v{model["version"]}')
print(f'Components: {list(model["components"].keys())}')
```

```python
# ── Explore persona constraints ──────────────────────────────────────
import pandas as pd

persona_df = pd.DataFrame(model['personas']).T
print('Customer Personas:')
persona_df
```

```python
# ── Product scorer details ───────────────────────────────────────────
scorer = model['components']['product_scorer']
print('Scoring Weights:')
for k, v in scorer['weights'].items():
    print(f'  {k}: {v} pts')

print(f'\nTotal possible: {sum(scorer["weights"].values())} pts')
```

```python
# ── Available customizations ─────────────────────────────────────────
custs = model['components']['customization_engine']['customizations']
cust_df = pd.DataFrame(custs)
print('Customization Options:')
cust_df
```
